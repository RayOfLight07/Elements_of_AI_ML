```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("visualization_data.csv")
```

```python
data = pd.read_csv("visualization_data.csv")
data.info()
```

```python
plt.figure()
plt.plot(data['spending_score'],data['is_loyal_customer'],marker='*')
plt.title('Line Plot')
plt.xlabel('spending_score')
plt.ylabel('is_loyal_customer')
plt.show()
```

```python
plt.figure()
plt.bar(data['age'],data['income'])
plt.title('Bar Chart')
plt.xlabel('age')
plt.ylabel('income')
plt.show()
```

```python
plt.figure()
plt.scatter(data['age'],data['spending_score'])
plt.title('Scatter Plot')
plt.xlabel('age')
plt.ylabel('spending_score')
plt.show()
```

```python
plt.figure()
sns.histplot(data['income'],bins=6)
plt.title('Histogram')
plt.xlabel('income')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='age',y='spending_score',data=data)
plt.title('Box Plot')
plt.xlabel('age')
plt.ylabel('spending_score')
plt.show()
```

```python
af=data[['income','age']]
corr= af.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, mask=mask, cmap='viridis')
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='YlGnBu', annot_kws={'size': 12, 'weight': 'bold'}, cbar_kws={'orientation': 'horizontal'})
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0)
plt.show()
```

```python
sns.heatmap(corr, annot=True, linewidths=1.5, linecolor='white', cmap='Spectral')
plt.show()
```

```python
mask = np.zeros_like(corr)
mask[corr == 0] = True
sns.heatmap(corr, annot=True, cmap='YlOrBr', mask=mask)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='rocket', annot_kws={"bbox": {"boxstyle": "round,pad=0.3", "fc": "yellow", "ec": "black", "lw": 1}})
plt.show()
```
